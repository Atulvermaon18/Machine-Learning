```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
si = SentimentIntensityAnalyzer()
def check_sentiment(message):
    if type(message)== str:
        sentiment = si.polarity_scores(message)
        if 0 <= sentiment['compound'] <= 0.3: 
            return 'Neutral'
        elif sentiment['compound'] >=0.3:
            return 'Positive'
        else:
            return 'Negative'
```

```python
product=['iPhone 8','iPhone X', 'iPhone XR', 'iPhone 11', 'iPhone 11pro']
```

```python
def add_product(n):
    random = np.random.randint(0,len(product))
    return product[random]
```

```python
yelp = pd.read_csv('yelp.csv')
```

```python
yelp['product'] = yelp['stars'].apply(add_product)
```

```python
yelp = yelp[['stars', 'text','product']]
```

```python
yelp
```

```python
yelp.info()
```

```python
yelp['prediction'] = yelp['text'].apply(check_sentiment)
```

```python
yelp.head()
```

```python
def eval(stars,pred):    
    if (stars == 1) & (pred=='Positive'):
        return 'Wrong Prediction'
    elif (stars == 2 )& (pred=='Positive'):
        return 'Wrong Prediction'
    else:
        return 'Correct Prediction'
```

```python
sns.countplot(yelp['prediction'])
```

```python
yelp['Eval'] = yelp.apply(lambda x: eval(x['stars'],x['prediction']),axis=1)    
```

```python
sns.countplot(yelp['Eval'])
```

```python
def show_report(prod):
    print('Product',prod)
    apple_prod = yelp[yelp['product']==prod]
    sns.countplot(apple_prod['prediction'])
```

```python
show_report(prod)
```
