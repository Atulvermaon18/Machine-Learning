```python
si = SentimentIntensityAnalyzer()
def check_sentiment(message):
    if type(message)== str:
        sentiment = si.polarity_scores(message)
        print(sentiment)
        if 0 <= sentiment['compound'] <= 0.3: 
            return 'Neutral'
        elif sentiment['compound'] >=0.3:
            return 'Positive'
        else:
            return 'Negative'
```

```python
check_sentiment("A Student said guru analogy stupid")
```
